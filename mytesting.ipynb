```python
import pandas as pd
import math
import os
```

```python
real_result = pd.read_csv("data/game_origin_results.csv")
real_result.head()
```

```python
our_result = pd.read_csv("output/game_winner_summary.csv")
our_result = our_result.drop(columns=["team_type"])
our_result = our_result.drop_duplicates()
our_result.head()
```

```python
joined_result = pd.merge(real_result, our_result, on='game_id')
joined_result.head()
joined_result[['game_id','home_points','away_points','prob_home_low','prob_home_high','prob_away_low','prob_away_high','home_expected_score','away_expected_score']]
```

```python
joined_result['real_winner'] = ['home' if x > y else 'away' for x, y in zip(joined_result['home_points'], joined_result['away_points'])]
joined_result.head()
```

```python
joined_result['my_winner'] = ['home' if x > y else 'away' for x, y in zip(joined_result['home_expected_score'], joined_result['away_expected_score'])]
joined_result['my_winner']
```

```python
joined_result[['game_id','home_points','away_points','real_winner','winner','my_winner']]
```

```python
percent_winnner_correct = ( (joined_result['real_winner']==joined_result['winner']).sum() )/ len(joined_result) * 100
percent_my_winnner_correct = ( (joined_result['real_winner']==joined_result['my_winner']).sum() )/ len(joined_result) * 100
print(f"{percent_winnner_correct}, {percent_my_winnner_correct}")
```

```python
percent_random_correct = ( (joined_result['winner']==joined_result['my_winner']).sum() )/ len(joined_result) * 100
percent_random_correct
```
